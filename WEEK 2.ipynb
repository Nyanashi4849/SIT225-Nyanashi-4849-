```python
import random
import matplotlib.pyplot as plt

n_values = 100
y_values = []

# Create data (y_values) randomly between 1 and 100.
for i in range(n_values):
    y_values.append(random.randint(1, 100))

x_values = range(n_values)  # X is sequence of values 0-99
plt.plot(x_values, y_values)
plt.show()
```

```python
# Plot 2 variables
# 

n_values = 100
y_values_1 = []
y_values_2 = []

# Create data (y_values) randomly between 1 and 100.
for i in range(n_values):
    y_values_1.append(random.randint(1, 100))
    y_values_2.append(random.randint(1, 100))

x_values = range(n_values)  # X is sequence of values 0-99
plt.plot(x_values, y_values_1)
plt.plot(x_values, y_values_2)  # call plot again draws in the same graph.
plt.show()
```

```python
import matplotlib.pyplot as plt

n_values = 500
y_values_1 = []

# Create ascending data for y_values_1
prev_value_1 = 1  # Start at 1
for i in range(n_values):
    prev_value_1 += 5  # Increment by 5 (or any fixed step)
    y_values_1.append(prev_value_1)

x_values = range(n_values)  # X is sequence of values 0-99
plt.plot(x_values, y_values_1, label="Line 1 (Increment by 5)")
plt.legend()
```

```python
import matplotlib.pyplot as plt

n_values = 500
y_values_1 = []

# Create decending data for y_values_1
prev_value_1 = 1  # Start at 1
for i in range(n_values):
    prev_value_1 -= 5  # decrement by 5 (or any fixed step)
    y_values_1.append(prev_value_1)

x_values = range(n_values)  # X is sequence of values 0-99
plt.plot(x_values, y_values_1, label="Line 1 (decrement by 5)")
plt.legend()
```

```python
import math
import matplotlib.pyplot as plt

n_values = 200
x_values = range(n_values)  # X values will be from 0 to 99

# Create wave-like data using the sin function
y_values_wave = [math.cos(0.25 * x) for x in x_values]  # 0.5 * x to control the wave frequency

plt.plot(x_values, y_values_wave, label="cos Wave")
plt.legend()
plt.show()
```
